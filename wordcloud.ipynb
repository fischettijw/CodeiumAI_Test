# Word Cloud - normal

```python
import matplotlib.pyplot as plt # type: ignore
from wordcloud import WordCloud # type: ignore

# Sample text
text = """
Python is an interpreted high-lel general-purpose programming language.
Its design philosophy emphasizes code readability with the use of significant indentation.
Its language constructs as well as its object-oriented approach aim to help programmers write clear, logical code for small and large-scale projects.
"""

# Generate the word cloud
# wordcloud = WordCloud().generate(text)
wordcloud = WordCloud(width=800, height=400, background_color="white").generate(text)

# Display the word cloud using matplotlib
plt.figure(figsize=(10, 5))

plt.axis("on")  # Show the axes
# plt.axis("off")  # Hide the axes

plt.imshow(wordcloud, interpolation="bilinear")
# plt.show()
```

# Word Cloud - random color palette

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import random

# Sample text
text = """
Python is an interpreted high-lel general-purpose programming language.
Its design philosophy emphasizes code readability with the use of significant indentation.
Its language constructs as well as its object-oriented approach aim to help programmers write clear, logical code for small and large-scale projects.
"""

# Custom color function
def custom_color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    return "hsl({}, {}%, {}%)".format(random.randint(0, 255), random.randint(60, 100), random.randint(30, 70))

# Generate the word cloud with the custom color function
wordcloud = WordCloud(width=800, height=400, background_color='white', color_func=custom_color_func).generate(text)

# Display the word cloud using matplotlib
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # Hide the axes
plt.show()
```

# px.scatter_geo

```python
import plotly.express as px
import pandas as pd

# Example DataFrame
data = {
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix'],
    'Latitude': [40.7128, 34.0522, 41.8781, 29.7604, 33.4484],
    'Longitude': [-74.0060, -118.2437, -87.6298, -95.3698, -112.0740],
    'Population': [8419000, 3980000, 2716000, 2328000, 1690000]
}

df = pd.DataFrame(data)

# Create the scatter_geo plot
fig = px.scatter_geo(
    data_frame= df,
    lat='Latitude',
    lon='Longitude',
    # text='City',  # Column to display as text
    # size='Population',
    projection='albers usa',    # simple 2D map projection
    title= 'Scatter Geo Plot Title',
    center={'lat': 37.0902, 'lon': -95.7129},
    width=800,   # Set the width of the figure
    height=600,   # Set the height of the figure
)

# Update text position
# fig.update_traces(textposition='top center')
fig.update_traces(textposition='top center', textfont=dict(size=12, color='black'))

# Update the layout to center the title
fig.update_layout(
    title={
        'text': 'Scatter Geo Plot Centered on the USA',  # Title text
        'x': 0.5,  # Center the title (0.5 is centered)
        'xanchor': 'center',  # Set the x anchor to the center
        'yanchor': 'top'  # Set the y anchor to the top
    })


# Show the plot
fig.show()
```
